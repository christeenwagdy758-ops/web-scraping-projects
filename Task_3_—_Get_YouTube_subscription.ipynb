```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.firefox import GeckoDriverManager
import time
profile_path = r"C:\Users\chris\AppData\Roaming\Mozilla\Firefox\Profiles\y9621c3t.default-release"
options = Options()
options.profile = profile_path

driver = webdriver.Firefox(
    service=Service(GeckoDriverManager().install()),
    options=options)
url="https://www.youtube.com/feed/channels"
driver.get(url)
time.sleep(15)
print(driver.title)
```

```python
from bs4 import BeautifulSoup
html=driver.page_source
soup = BeautifulSoup(html, "html.parser")
channels=soup.find_all("div",class_="style-scope ytd-channel-renderer")
seen = set()
time.sleep(15)
for c in channels:
    name_channel = c.find("yt-formatted-string", id="text")
    num_sup = c.find("span", id="video-count")
    if not (name_channel and num_sup ):
        continue

    name_channel_text =name_channel.get_text(strip=True)
    num_sup_text = num_sup.get_text(strip=True)
    if name_channel_text in seen:
                continue
    seen.add(name_channel_text)


    print( name_channel_text, "|",num_sup_text)
```

```python
import csv
with open("youtube_sup.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Name Channel", "Number of sub"])
    for c in channels:
        try:
            name_channel = c.find("yt-formatted-string", id="text")
            num_sup = c.find("span", id="video-count")
            if not (name_channel and  num_sup ):
                continue

            name_channel_text =   name_channel.get_text(strip=True)
            num_sup_text =  num_sup.get_text(strip=True)
            if name_channel_text in seen:
                continue
            seen.add(name_channel_text)
            writer.writerow([name_channel_text, num_sup_text])
        except Exception:
            continue

driver.quit()
print("DONE")
```
