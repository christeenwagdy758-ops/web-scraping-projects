```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.firefox import GeckoDriverManager
import time
options = webdriver.FirefoxOptions()
driver = webdriver.Firefox(service=Service( GeckoDriverManager().install()), options=options)
url="https://www.carrefouregypt.com/mafegy/en/c/NFEGY4070500"
driver.get(url)
time.sleep(15)
print(driver.title)
```

```python
from bs4 import BeautifulSoup
html=driver.page_source
soup = BeautifulSoup(html, "html.parser")
products=soup.find_all("div",class_="relative")
time.sleep(15)
for p in products:
    name = p.find("div", class_="text-sm leading-4 font-medium line-clamp-2 text-left md:text-md")
    price = p.find("div", class_="mr-2xs flex w-fit items-center force-ltr text-c4blue-800")
    link = p.find("a", href=True)
    if not (name and price and link):
        continue

    name_text = name.get_text(strip=True)
    price_text = price.get_text(strip=True)
    link_text = "https://www.carrefouregypt.com" + link['href']

    print(name_text, "|", price_text, "|", link_text)
```

```python
import csv
with open("carrefour_producttry1.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Product Name", "Price (EGP)", "link"])
    for p in products:
        try:
            name = p.find("div", class_="text-sm leading-4 font-medium line-clamp-2 text-left md:text-md")
            price = p.find("div", class_="mr-2xs flex w-fit items-center force-ltr text-c4blue-800")
            link = p.find("a", href=True)
            if not (name and price and link):
                 continue

            name_text = name.get_text(strip=True)
            price_text = price.get_text(strip=True)
            link_text = "https://www.carrefouregypt.com" + link['href']
            writer.writerow([name_text, price_text, link_text])
        except Exception:
            continue

driver.quit()
print("DONE")
```
